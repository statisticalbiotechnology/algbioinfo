```python
from matplotlib import pyplot as plt
from pylab import *
t=np.arange(.0,2,.01)
s=sin(2.*pi*t)
c=cos(2*pi*t)
plt.plot(t,s)
plt.show()
```

```python
plt.plot(t,s,'.',color='green', label="sin(x)")
plt.plot(t,c,'.',color='blue', label="cos(x)")
plt.xlabel("Angle")
plt.ylabel("sin/cos")
plt.legend(loc="upper left")
plt.show()
```

```python
import seaborn as sns
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
tips["tip_rate"] = tips["tip"]/(tips["tip"]+tips["total_bill"])
```

```python
sns.catplot(x="day",y="tip", hue="size",data=tips[tips["sex"]=="Female"], kind="box")
```

```python
sns.catplot(x="size",y="tip",data=tips, kind="swarm")
```

```python
sns.catplot(x="size",y="total_bill",data=tips, kind="swarm")
```

```python
tips["tip_rate"] = tips["tip"]/tips["total_bill"]
sns.lmplot(x="total_bill",y="tip_rate",hue="smoker",data=tips,fit_reg=True)
```

```python
sns.catplot(x="day",y="tip", hue="sex",data=tips[tips["tip"]>5], kind="swarm")
```

```python
sns.catplot(x="day",y="tip", hue="sex",data=tips[tips["tip"]<1.5], kind="swarm")
```

```python
tips["tip_p_size"] = tips["tip"]/tips["size"]
tips["bill_p_size"] = tips["total_bill"]/tips["size"]
sns.lmplot(x="bill_p_size",y="tip_p_size",data=tips,col="size", 
           hue="size",col_wrap=2)
```

```python
sns.catplot(x="size",y="tip_rate",data=tips,hue="size",kind="box")
```
